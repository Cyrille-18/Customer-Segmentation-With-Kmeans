## 03-Feature Engineering

The objective of this step is to create more meaningful variables from the cleaned dataset. `../data/processed/customer_cleaned.csv`

These new features will improve both the exploratory analysis and customer segmentation process.

```python
import pandas as pd
from src.feature_engineering import create_features
```

### 1. Load cleaned data
The cleaned dataset is loaded with `Dt_Customer` parsed as datetime directly on read,
so no conversion is needed later.

```python
# load data
cleaned_data = pd.read_csv("../data/processed/customer_cleaned.csv",parse_dates=['Dt_Customer'])
cleaned_data.head(5)
```

### 2. Feature engineering

The `create_features()` function builds the following derived columns :

| Feature | Description |
|---|---|
| `Age` | Customer age derived from `Year_Birth` |
| `Total_Children` | Sum of `Kidhome` and `Teenhome` |
| `Total_Spending` | Total spend across all product categories |
| `Customer_since` | Number of days since enrollment |
| `accepted_campaigns` | 1 if the customer responded to at least one campaign |
| `Age_Group` | Age bracket (18-29, 30-39, ..., 70+) |

```python
cleaned_data = create_features(cleaned_data)
```

Preview the changes

```python
cleaned_data.head(5)
```

### 3. Export

The enriched dataset is exported to `data/processed/` for use in the next step.

```python
cleaned_data.to_csv('../data/processed/customer_features.csv', index=False)
```
